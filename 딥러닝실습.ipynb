```python
# X, y 분리
X = data.iloc[:,:-1]
y = data.iloc[:,-1]

X.shape, y.shape
```

```python
# 훈련데이터 / 테스트 데이터 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=7)
```

### 신경망 설계

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model1 = Sequential()

# 입력 층
model1.add(Dense(32,input_dim= X_train.shape[1], activation="relu")) #특성 개수(17)

# 은닉 층
model1.add(Dense(64, activation="relu"))
model1.add(Dense(32, activation="relu"))
model1.add(Dense(16, activation="relu"))

# 출력 층
model1.add(Dense(1, activation="sigmoid"))

model1.summary()
```

```python
#모델 컴파일
model1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["acc"])
```

```python
model1.fit(X_train,y_train, epochs=500, batch_size=50)
```

```python
# 평가
print("훈련 데이터 :", model1.evaluate(X_train,y_train))
print("테스트 데이터 :", model1.evaluate(X_test,y_test))
```

## Wine 데이터를 이용한 딥러닝 개인 실습

```python
# X, y 분리

# type 컬럼을 이용한 이진분류
X1 = wine.iloc[:,:-1]
y1 = wine.iloc[:,-1]

# quality 컬럼을 이용한 다중분류
X2= wine.drop("quality",axis=1)
y2 = wine.loc[:,"quality"]

# alcohol 컬럼을 이용한 회귀
X3 = wine.drop("alcohol",axis=1)
y3 = wine.loc[:,"alcohol"]
```
