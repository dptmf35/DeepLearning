```python
import pandas as pd
import numpy as np
```

```python
# AND 논리를 이용한 신경망
XData = [[0,0], [0,1],[1,0],[1,1]]
y = [0,0,0,1]
```

```python
# w와 b를 초기화 (항상 같은값으로 w,b가 초기화되도록 함)
seed = 7
np.random.seed(seed)
tensorflow.random.set_seed(seed)
```

### 신경망 설계

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

# Sequential : 신경망을 구성해주는 기능
# Dense : 신경망 층 설정
# Activation : 활성화 함수를 설정
```

```python

model = Sequential()

# 입력층
# add() : 신경망에 신경망 층을 추가
# 노드의 출력 수 , 노드의 입력 수 (특성 수)
# input_dim : 입력 층의 개수 (특성수)
model.add(Dense( 6, input_dim = 2))
#model.add(Activation("relu"))

# 출력층
# 회귀 : linear (안써도 됨)
# 이진 분류 : sigmoid, softmax
# 다중분류 : softmax
model.add(Dense(1))
model.add(Activation("sigmoid"))
# summary() : 신경망 구조를 표시
model.summary()
```

```python
# compile() : 학습이 가능한 모델로 변환
# loss : 손실함수
#   - 회귀 : mse (또는 mean squared error)
#   - 이진분류 : binary crossentropy
#   - 다중분류 : categorical crossentropy
# optimizer : 최적화함수(경사하강법) - adam
# metrics : 평가방법 : acc 또는 accuracy

model.compile(loss="binary_crossentropy", optimizer ="adam", metrics=["acc"])
```

```python
# 학습
# 특성데이터(X), 라벨데잍(y)
# epochs : w, b를 몇 번 갱신할 것인지?(몇 번 반복?)
# batch_size : 한 번 반복할 때 몇 개의 데이터를 이용할 것인지
model.fit(XData, y, epochs=1000, batch_size = 4)
```

### 논리 퍼셉트론 직접 구현해 보자!

AND 논리

```python
# 입력 파라미터가 2개이고 출력 파라미터(반환값)가 1개인 함수
def AND_perceptron(x1, x2) :
    # 파라미터 초기화(w1, w2, b)
    w1, w2, b = 0.5, 0.3, -0.3
    
    temp = w1 * x1 + w2 * x2 + b
    
    if temp >= 0.5 :
        return 1
    else :
        return 0
    
```

```python
print(AND_perceptron(0,0))
print(AND_perceptron(0,1))
print(AND_perceptron(1,0))
print(AND_perceptron(1,1))
```

```python
def OR_perceptron(x1, x2) :
    # 파라미터 초기화(w1, w2, b)
    w1, w2, b = 0.7, 0.2, 0.4
    
    temp = w1 * x1 + w2 * x2 + b
    
    if temp >= 0.5 :
        return 1
    else :
        return 0
```

```python
print(OR_perceptron(0,0))
print(OR_perceptron(1,0))
print(OR_perceptron(0,1))
print(OR_perceptron(1,1))
```

```python
def NAND_perceptron(x1, x2) :
    # 파라미터 초기화(w1, w2, b)
    w1, w2, b = -0.1, -0.1, 0.7
    
    temp = w1 * x1 + w2 * x2 + b
    
    if temp >= 0.5 :
        return 1
    else :
        return 0
```

```python
print(NAND_perceptron(0,0))
print(NAND_perceptron(1,0))
print(NAND_perceptron(0,1))
print(NAND_perceptron(1,1))
```

```python
def XOR_perceptron(x1, x2) :
    s1 = NAND_perceptron(x1, x2)
    s2 = OR_perceptron(x1, x2)
    y = AND_perceptron(s1, s2)
    
    return y
```

```python
print(XOR_perceptron(0,0))
print(XOR_perceptron(1,0))
print(XOR_perceptron(0,1))
print(XOR_perceptron(1,1))
```

### 오차역전파 구현

```python
#1개의 노드로 구성된 1층 신경망
import numpy as np

#항상 같은 값으로 초기화
np.random.seed(0)

def backpropagation1(X,epochs, lr) : # 특징값, 반복수, 학습률
    # 1. w, b를 초기화
    w = np.random.rand(1) # 0 ~ 1 값 중 한개를 추출
    b = np.random.rand(1)
    
    #2. epochs 만큼 반복해서 학습
    for i in range(epochs) :
        # 2-1. 첫번째 반복에서 X를 입력하여 y를 구함
        pred_y = w * X + b
        
        # 2-2. 오차계산
        error = X - pred_y
        
        print("반복 수 : {} , 예측 값 : {}, 오차: {}".format(i, pred_y, error))
        
        # 2-3. 오차 역전파
        # 기울기 : 오차 / 예측값 (미분)
        d1 = error / pred_y
        
        # w와 b를 갱신
        # lr : 학습률(w,b를 갱신하는 데 기울기 비율을 어떻게 적용할 것인지?)
        w = w + lr * d1
        b = b + lr * d1
        
```

```python
backpropagation1(2, 30, 0.01)
```

```python
#2개의 노드로 구성된 2개의 신경망

np.random.seed(0)

def backpropagation2(X, epochs, lr) :
    # 1. w, b 초기화
    w1 = np.random.rand(1)
    w2 = np.random.rand(1)    
    b1 = np.random.rand(1)    
    b2 = np.random.rand(1)
    
    for i in range(epochs) :
        # 2. 층별 예측값
        pred_h = w1 * X + b1
        pred_y = w2 * pred_h + b2
        
        # 3. 2층 오차 계산
        error = X - pred_y
        
        print("반복 수 : {} , 예측 값 : {}, 오차: {}".format(i, pred_y, error))
        
        # 4. 2층 오차 역전파
        # 오차(error) / 예측값 (미분)
        d2 = error / pred_y
        
        # 5. w2, b2 갱신
        w2 = w2 + lr * d2
        b2 = b2 + lr * d2
        
        # 6. 1층 오차 역전파
        d1 = pred_y / pred_h * d2
        
        # 7. w1, b1 갱신
        w1 = w1 + lr * d1
        b1 = b1 + lr * d1        
```

```python
backpropagation2(2, 30, 0.01)
```

### 은닉층

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

# Sequential : 신경망을 구성해주는 기능
# Dense : 신경망 층 설정
# Activation : 활성화 함수를 설정

model = Sequential()

# 입력층
# add() : 신경망에 신경망 층을 추가
# 노드의 출력 수 , 노드의 입력 수 (특성 수)
# input_dim : 입력 층의 개수 (특성수)
model.add(Dense( 4, input_dim = 2))
model.add(Activation("relu"))

#은닉층(데이터분석 - 특성추출)
model.add(Dense(8, activation="relu"))

# 출력층
model.add(Dense(1))
model.add(Activation("sigmoid"))
# summary() : 신경망 구조를 표시
model.summary()
```

```python
XData = [[0,0], [0,1],[1,0],[1,1]]
y = [0,0,0,1]
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["acc"])
model.fit(XData,y, epochs=100, batch_size=4)
```
